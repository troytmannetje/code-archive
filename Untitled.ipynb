```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cm
from matplotlib import colormaps
```

#### Create data to plot

```python
def func(x1, x2=5, x3=3, x4=2, x5=1):
    y = np.sqrt(np.sqrt((x1*np.sin(x2-2)+x3*np.cos(x4))**2)+6*np.cos(x5)**3)
    return y
```

and create the axes:

```python
x1_ = np.linspace(0, 10,100)
x2_ = np.linspace(0, 15,100)
x3_ = np.linspace(-10, 10,11)
x4_ = np.linspace(-10, 10,11)
x5_ = np.linspace(-10, 10,11)
```

### 2-dimensions

```python
x1 = x1_
y = func(x1, 3 ,2,23,45)
print(x1.shape)
print("Done!")
```

#### Simple line plot

```python
fig, ax = plt.subplots(layout='constrained')

# plot datapoints
ax.plot(x1, y, label = "line 1", 
        color="black", alpha =0.8,
        linestyle="-",linewidth=2,
        marker=None,markersize=2)

# set plot area
ax.set_xlim([0,10])
ax.set_ylim([0,3])

# create labels
ax.legend()
ax.set_ylabel(f'$y$', rotation='horizontal', labelpad=15)
ax.set_xlabel(f'$x_1$')

# format ticks and gridlines
ax.minorticks_on()
ax.tick_params(axis='both', which='major', labelsize=8,
               direction='inout', length=4,
               bottom=True, left=True, right=True,
               labelleft=True, labelright=True,
               rotation=None)
ax.grid(which='major')
```

Visualise how much line 1 diverges from the mean value:

```python
# create and plot reference data
reference=y.mean()
ax.axhline(reference, color='red', linestyle=None)

# compare difference between two sets of datapoints
cmap = colormaps['seismic'] # select colormap
norm = colors.Normalize(vmin=-2,vmax=2) # normalisation method
errorcolor = cmap(norm(y-reference)) # convert float between 0 to 1 to RGB

# fill 
for i in range(len(x1)-1):
    ax.fill_between(x1[i:(i+2)], y[i:(i+2)], reference, interpolate=True,
                    facecolor=errorcolor[i],edgecolor='face')   
cbar = fig.colorbar(cm.ScalarMappable(norm=norm, cmap=cmap),ax=ax,aspect=40)
cbar.set_label(f"error")

fig
```

### 3D plots
First create points on every combination of axes:

```python
x1, x2 = np.meshgrid(x1_, x2_)
y = func(x1, x2,2,23,45)
print(x1.shape)
print("Done!")
```

#### Contour map

- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html
- https://matplotlib.org/stable/users/explain/colors/colormaps.html


```python
fig, ax = plt.subplots(layout='constrained')


# plot colour gradient
fill_levels = np.linspace(1,y.max(),100) #list of levels to plot (linspaced if int)
norm = colors.Normalize(fill_levels.min(),fill_levels.max()) #normalisation method to change value to color

contourfill = ax.contourf(x1,x2,y,levels= fill_levels,
                          cmap=plt.cm.viridis, norm=norm,
                          extend="both")
contourfill.cmap.set_under("black") # color if value < min level
contourfill.cmap.set_over("white") # color if value > max level

heightmarks = np.arange(fill_levels.min(),fill_levels.max(),0.2).round(1) # 
contourline = ax.contour(x1,x2,y,levels = heightmarks, 
                         colors=["black"],linestyles=["-"],linewidths=0.5,
                          extend="both")
ax.clabel(contourline, colors='black', fontsize=5) 

cbar = fig.colorbar(contourfill,ticks=heightmarks[::5],aspect=40)
cbar.add_lines(contourline)

ax.set_xlabel(f"$x_1$")
ax.set_ylabel(f"$x_2$")
cbar.set_label(f"$y$")
```
